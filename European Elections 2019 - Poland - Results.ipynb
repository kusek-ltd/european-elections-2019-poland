````python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import geopandas as gpd

import os

projectDirectory = "/Users/nataliakusek/PycharmProjects/european-elections-2019-poland"
````

````python
obreby = gpd.read_file(os.path.join(projectDirectory, 'jednostki_administracyjne', 'obreby_crs.shp'))
````

````python
gminy = gpd.read_file(os.path.join(projectDirectory, 'jednostki_administracyjne', 'gminy_crs.shp'))
````

````python
fig, ax = plt.subplots(1, figsize=(20, 14))
obreby.plot(ax=ax, linewidth=0.5, edgecolor='0.4', color='0.8')
ax.axis('off')
````

````python
fig, ax = plt.subplots(1, figsize=(14, 14))
gminy.plot(ax=ax, linewidth=0.5, edgecolor='0.4', color='0.8')
ax.axis('off')
````
